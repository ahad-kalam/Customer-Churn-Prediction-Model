```python
df1_TotalCharges = pd.to_numeric(df1.TotalCharges)
df1_TotalCharges.dtype
```

```python
df2 = df1[['tenure','Churn']]
```

```python
df2_churn_yes = df2[df2['Churn'] == 'Yes'].tenure
df2_churn_yes.head()
```

```python
df2_churn_no = df2[df2['Churn'] == "No"].tenure
df2_churn_no.head()
```

```python
plt.hist([df2_churn_yes,df2_churn_no],color=['green','red'],label=['Churn=Yes','Churn=No'])
plt.legend()
plt.xlabel('Tenure in Month')
plt.ylabel('Number of Customers')
plt.title('Customer Churn Prediction Vistualiztion')
```

```python
mc_churn_yes = df1[df1['Churn']=='Yes'].MonthlyCharges
mc_churn_yes.head()
```

```python
plt.hist([mc_churn_yes,mc_churn_no],color=['green','orange'],label=['Churn=Yes','Churn=No'])
plt.legend()
plt.xlabel('Monthly Charges in Rupees')
plt.ylabel('Number of Customer')
plt.title('Churn Prediciton Monthly Charges')
```
